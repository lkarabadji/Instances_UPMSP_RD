```python
import pandas as pd
data = pd.read_excel('best_results_Rm_ri_Sijk_Cmax_2023.xlsx')
```

```python
data.head(10)
```

```python
data.shape
```

```python
# Sélection des caractéristiques pertinentes
features = ['N', 'M', 'Setup', 'Release', 'Best_cmax', 'SA_cmax','SA_runtime', 'lahc_cmax', 'lahc_runtime']
```

```python
# Création de la matrice de caractéristiques
X = data[features]
```

```python
X
```

```python
#Définir le nombre de clusters
n_clusters = 100
```

```python
import pandas as pd
from sklearn.cluster import KMeans
# Application de l'algorithme de clustering k-means
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

```python
# Define a function to convert intervals to numerical values
def convert_interval_to_value(interval):
    if isinstance(interval, str) and '-' in interval:
        a, b = interval.split('-')
        return (int(a) + int(b)) / 2
    else:
        return interval
```

```python
X['Setup'] = X['Setup'].apply(convert_interval_to_value)
```

```python
X
```

```python
X = X.drop(index=X.index[-1])
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

```python
# Obtention des centroïdes pour chaque cluster
centroids = kmeans.cluster_centers_
```

```python
# Obtention des étiquettes de cluster pour chaque instance
labels = kmeans.labels_
```

```python
import numpy as np
representatives = []
for i in range(n_clusters):
    # Obtention de l'indice des instances appartenant à ce cluster
    idx = [j for j in range(len(labels)) if labels[j] == i]
    
    # Calcul de la distance entre chaque instance et le centroïde
    distances = [np.linalg.norm(X.iloc[j,:] - centroids[i,:]) for j in idx]
    
    # Obtention de l'instance la plus proche du centroïde
    representative_idx = idx[np.argmin(distances)]
    representatives.append(representative_idx)

# Sélection des instances représentatives
representatives_df = data.iloc[representatives,:]
```
